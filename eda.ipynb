Do your EDA in this notebook!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mh_1 = pd.read_csv('data/mentalillness_data1.csv')
mh_2 = pd.read_csv('data/mentalillness_data2.csv')
mh = mh_1.append(mh_2)
mh
```

```python
mh.shape
```

```python
mh.describe()
```

```python
mh.dropna(inplace = True)
```

```python
mh
```

```python
mh_by_state = mh.groupby(['state','illness']) \
                .count() 
mh_by_state
```

```python
mh['illness'].value_counts().plot(kind='bar')
plt.ylabel('Number of Cases')
plt.xlabel('Disorder')
plt.title('Number of Cases by Disorder');
```

```python
mh['gender'].value_counts().plot(kind='bar')
plt.ylabel('Number of Cases')
plt.xlabel('Gender')
plt.title('Number of Cases by Gender');
```

```python
mh['ethnicity'].value_counts().plot(kind='bar')
plt.ylabel('Number of Cases')
plt.xlabel('Ethnicity')
plt.title('Number of Cases by Ethnicity');
```

```python
mh['state'].value_counts().plot(kind='bar', x='State', y='Total Cases', figsize=(13,8));
```

```python
print(mh.dtypes)
```

```python
mh_by_gender = mh.groupby(['gender','illness']) \
                .count()

mh_by_gender
```

```python
mh_by_ethnicity = mh.groupby(['ethnicity','illness']) \
                .count() \
                .sort_values('illness', ascending=False)

mh_by_ethnicity
```
